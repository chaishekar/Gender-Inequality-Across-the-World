```python
import pandas as pd
import numpy as np
```

```python
# read the csv file
file_path = "../data/raw-data/gender/gender index data/Gender_Inequality_Index.csv"
# Load the CSV file into a DataFrame
df = pd.read_csv(file_path)
df.head()
```

```python
# clean the data
# check nan values
df.isnull().sum()

# fill nan values as Not known
df.fillna("Not known", inplace=True)

# remove year row
df.drop([0], inplace=True)

# check for column names
df.columns

# Rename the columns using the rename() method
df = df.rename(columns={
    'HDI rank': 'HDI Rank',
    'Gender Inequality Index.Value': 'Gender Inequality Index Value',
    'Gender Inequality Index.Rank': 'Gender Inequality Index Rank',
    'Maternal mortality ratio(deaths per 100,000 live births)': 'Maternal Mortality Ratio',
    'Adolescent birth rate(births per 1,000 women ages 15–19)': 'Adolescent Birth Rate',
    'Share of seats in parliament(% held by women)': 'Women Share of Seats in Parliament ',
    'Population with at least some secondary education.Female': 'Female Population with Some Secondary Education',
    'Population with at least some secondary education.Male': 'Male Population with Some Secondary Education',
    'Labour force participation rate .Female': ' Female Labour Force Participation Rate',
    'Labour force participation rate .Male': 'Male Labour Force Participation Rate'
})

df.head()
```

```python
# save to csv file 
df.to_csv('../data/clean_data/clean_gender_inequality_index.csv', index=False)
```
